```python
# Import the tools we need
from PIL import Image
import matplotlib.pyplot as plt

# Import our new FastEditor class from the other file
from src.edit_utils import FastEditor

# --- SETUP ---
# 1. Create an instance of our editor
editor = FastEditor()

# 2. Define the path to our test image
image_path = "data/test_images/elephant.jpg" 

# 3. Open the image
original_image = Image.open(image_path)

# --- ACTION ---
# Use our new method to make the image 50% brighter
bright_image = editor.adjust_brightness(original_image, factor=1.5)

# --- VERIFICATION ---
# Display the images side-by-side to compare
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(original_image)
ax[0].set_title("Original")
ax[0].axis('off') # Hide the axes

ax[1].imshow(bright_image)
ax[1].set_title("Brighter (Factor: 5)")
ax[1].axis('off')

plt.show()
```

```python
# --- ACTION (Contrast Test) ---
# Use our new method to make the image have 50% more contrast
contrast_image = editor.adjust_contrast(original_image, factor=1.5)

# --- VERIFICATION ---
# Display the images side-by-side to compare
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(original_image)
ax[0].set_title("Original")
ax[0].axis('off')

ax[1].imshow(contrast_image)
ax[1].set_title("More Contrast (Factor: 1.5)")
ax[1].axis('off')

plt.show()
```

```python
# --- ACTION (Saturation Test) ---
# Use our new method to make the image super vibrant (200% saturation)
saturated_image = editor.adjust_saturation(original_image, factor=2.0)

# --- VERIFICATION ---
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(original_image)
ax[0].set_title("Original")
ax[0].axis('off')

ax[1].imshow(saturated_image)
ax[1].set_title("More Saturated (Factor: 2.0)")
ax[1].axis('off')

plt.show()
```

```python
# --- ACTION (Blur Test) ---
# Use our new method to apply a blur with a radius of 5
blurred_image = editor.apply_blur(original_image, radius=5)

# --- VERIFICATION ---
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(original_image)
ax[0].set_title("Original")
ax[0].axis('off')

ax[1].imshow(blurred_image)
ax[1].set_title("Blurred (Radius: 10)")
ax[1].axis('off')

plt.show()
```

```python
# --- ACTION (Rotate Test) ---
# Use our new method to rotate the image by 30 degrees
rotated_image = editor.rotate(original_image, angle=30)

# --- VERIFICATION ---
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(original_image)
ax[0].set_title("Original")
ax[0].axis('off')

ax[1].imshow(rotated_image)
ax[1].set_title("Rotated by 30°")
ax[1].axis('off')

plt.show()
```

```python
# --- ACTION (Resize Test) ---
# Use our new method to resize the image to be 256x256 pixels
new_size = (256, 256)
resized_image = editor.resize(original_image, size=new_size)

# --- VERIFICATION ---
# Displaying the resized image by itself
print(f"Original size: {original_image.size}")
print(f"New size: {resized_image.size}")

plt.imshow(resized_image)
plt.title(f"Resized to {new_size}")
plt.axis('off')
plt.show()
```
